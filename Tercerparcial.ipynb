# Problema 1. Línea de espera con dos servidores en serie

Se implementa el seudocódigo propuesto por Sheldon Ross.

## Datos

$$
\lambda = 2
$$

$$
\mu_1 = 3
$$

$$
\mu_2 = 4
$$

$$
T = 100
$$

## Variables

$$
t = \text{tiempo}
$$

$$
N_A = \text{número de llegadas}
$$

$$
N_D = \text{número de salidas}
$$

$$
n_1 = \text{clientes en el servidor 1}
$$

$$
n_2 = \text{clientes en el servidor 2}
$$

## Eventos

$$
t_A = \text{siguiente llegada}
$$

$$
t_1 = \text{fin de servicio en el servidor 1}
$$

$$
t_2 = \text{fin de servicio en el servidor 2}
$$

## Distribuciones

$$
X=-\frac{1}{\lambda}\ln(U)
$$

$$
Y_1=-\frac{1}{\mu_1}\ln(U)
$$

$$
Y_2=-\frac{1}{\mu_2}\ln(U)
$$

$$
U\sim U(0,1)
$$


```python
import numpy as np
import random as r
import math as m
import matplotlib.pyplot as plt
```

```python
def exp(tasa):
    U = r.random()
    return -(1/tasa)*np.log(U)
```

```python
def SLE_2_servidores(lam, mu1, mu2, T):

    t = 0
    NA = 0
    ND = 0

    n1 = 0
    n2 = 0

    A1 = []
    A2 = []
    D = []

    S1 = [0]
    S2 = [0]
    ST = [0]
    tiempo = [0]

    tA = exp(lam)
    t1 = m.inf
    t2 = m.inf

    while True:

        # Caso 1: ocurre una llegada
        if tA <= t1 and tA <= t2 and tA <= T:

            t = tA
            NA = NA + 1
            n1 = n1 + 1

            A1.append(t)

            tA = t + exp(lam)

            if n1 == 1:
                t1 = t + exp(mu1)

            S1.append(n1)
            S2.append(n2)
            ST.append(n1+n2)
            tiempo.append(t)

        # Caso 2: termina servicio en servidor 1
        elif t1 < tA and t1 <= t2 and t1 <= T:

            t = t1

            n1 = n1 - 1
            n2 = n2 + 1

            A2.append(t)

            if n1 == 0:
                t1 = m.inf
            else:
                t1 = t + exp(mu1)

            if n2 == 1:
                t2 = t + exp(mu2)

            S1.append(n1)
            S2.append(n2)
            ST.append(n1+n2)
            tiempo.append(t)

        # Caso 3: termina servicio en servidor 2
        elif t2 < tA and t2 < t1 and t2 <= T:

            t = t2

            ND = ND + 1
            n2 = n2 - 1

            D.append(t)

            if n2 == 0:
                t2 = m.inf
            else:
                t2 = t + exp(mu2)

            S1.append(n1)
            S2.append(n2)
            ST.append(n1+n2)
            tiempo.append(t)

        # Después de T, ya no aceptamos nuevas llegadas,
        # pero los clientes que quedan en el sistema terminan su servicio.
        elif min(t1, t2) > T and (n1+n2) == 0:
            Tp = max(t - T, 0)
            break

        elif n1+n2 > 0:

            if t1 <= t2:

                t = t1

                n1 = n1 - 1
                n2 = n2 + 1

                A2.append(t)

                if n1 == 0:
                    t1 = m.inf
                else:
                    t1 = t + exp(mu1)

                if n2 == 1:
                    t2 = t + exp(mu2)

            else:

                t = t2

                ND = ND + 1
                n2 = n2 - 1

                D.append(t)

                if n2 == 0:
                    t2 = m.inf
                else:
                    t2 = t + exp(mu2)

            S1.append(n1)
            S2.append(n2)
            ST.append(n1+n2)
            tiempo.append(t)

        else:
            Tp = max(t - T, 0)
            break

    A1 = np.array(A1)
    D = np.array(D)

    clientes_completos = min(len(A1), len(D))

    if clientes_completos > 0:
        W = D[:clientes_completos] - A1[:clientes_completos]
        tiempo_promedio = np.mean(W)
    else:
        W = []
        tiempo_promedio = 0

    return tiempo, S1, S2, ST, A1, A2, D, W, tiempo_promedio, Tp, NA, ND
```

```python
lam = 2
mu1 = 3
mu2 = 4
T = 100

Z = SLE_2_servidores(lam, mu1, mu2, T)
```

```python
plt.step(Z[0], Z[3])
plt.xlabel("Tiempo")
plt.ylabel("Clientes en el sistema")
plt.title("Línea de espera con dos servidores en serie")
plt.grid()
plt.show()
```

```python
plt.step(Z[0], Z[1], label="Servidor 1")
plt.step(Z[0], Z[2], label="Servidor 2")
plt.xlabel("Tiempo")
plt.ylabel("Clientes")
plt.title("Clientes en cada servidor")
plt.legend()
plt.grid()
plt.show()
```

```python
print("Número de llegadas:", Z[10])
print("Número de salidas:", Z[11])
print("Tiempo promedio en el sistema:", Z[8])
print("Tiempo posterior a T:", Z[9])
```

## Interpretación de resultados

El número de llegadas representa la cantidad total de clientes que entraron al sistema durante el tiempo de simulación \(T\).

El número de salidas representa la cantidad de clientes que terminaron el servicio en ambos servidores.

El tiempo promedio en el sistema se calcula como:

$$
W = D - A_1
$$

donde $$\(A_1\)$$ es el tiempo de llegada del cliente al sistema y \(D\) es el tiempo de salida después de terminar el servicio en el servidor 2.

El valor $$\(T_p\)$$ representa el tiempo posterior a \(T\) en el que sale el último cliente que todavía se encontraba dentro del sistema.

Como el sistema tiene dos servidores en serie, cada cliente debe pasar primero por el servidor 1 y después por el servidor 2. Por eso, la simulación registra tres tipos de eventos: llegadas, terminación del servicio en el servidor 1 y terminación del servicio en el servidor 2.

```python
resultados_W = []
resultados_ND = []

for i in range(100):
    Z = SLE_2_servidores(lam, mu1, mu2, T)
    resultados_W.append(Z[8])
    resultados_ND.append(Z[11])

print("Promedio del tiempo en el sistema:", np.mean(resultados_W))
print("Desviación estándar del tiempo en el sistema:", np.std(resultados_W))
print("Promedio de salidas:", np.mean(resultados_ND))
```

# Conclusión

Después de realizar varias simulaciones, se obtiene una estimación del tiempo promedio que un cliente permanece en el sistema y del número promedio de clientes que completan el servicio.

El procedimiento sigue la lógica del seudocódigo propuesto por Sheldon Ross, donde el tiempo avanza al siguiente evento más cercano.

Por lo tanto, la simulación permite representar correctamente una línea de espera con dos servidores en serie mediante un modelo de eventos discretos.

# Problema 2. Variables antitéticas

Se desea estimar la integral

$$
\int_0^1 e^{x^2}dx
$$

mediante simulación Monte Carlo y variables antitéticas.

Sea

$$
f(x)=e^{x^2}
$$

y

$$
U \sim U(0,1).
$$

En primer lugar se utiliza el estimador de Monte Carlo crudo basado en

$$
f(U).
$$

Posteriormente se emplea la variable antitética

$$
1-U,
$$

obteniendo el estimador

$$
f(1-U).
$$

Finalmente se combinan ambos valores mediante

$$
\frac{f(U)+f(1-U)}{2}.
$$

El objetivo es comparar la varianza de los estimadores y verificar la reducción de variabilidad obtenida con variables antitéticas.

```python
import numpy as np
import random as r
```

```python
def f(x):
    return np.exp(x**2)
```

```python
f(0.5)
```

## Monte Carlo crudo

Primero se estima la integral usando directamente valores aleatorios:

$$
U_i \sim U(0,1)
$$

y se calcula:

$$
X_1=\frac{1}{N}\sum_{i=1}^{N} f(U_i)
$$

```python
def mc1(f,N):
    s=0

    for i in range(N):
        U = r.random()
        s += f(U)

    return s/N
```

```python
l1 = []

for i in range(100):
    M = mc1(f,1000)
    l1.append(M)

print(np.mean(l1))
print(np.var(l1))
```

## Estimador usando \(f(1-U)\)

Después se calcula otro estimador usando la variable antitética:

$$
1-U_i
$$

Entonces:

$$
X_2=\frac{1}{N}\sum_{i=1}^{N} f(1-U_i)
$$

```python
def mc2(f,N):
    s=0

    for i in range(N):
        U = r.random()
        s += f(1-U)

    return s/N
```

```python
l2 = []

for i in range(100):
    M = mc2(f,1000)
    l2.append(M)

print(np.mean(l2))
print(np.var(l2))
```

## Estimador antitético

Ahora se combinan los dos estimadores anteriores.

Para cada valor de \(U_i\), se utiliza también \(1-U_i\). De esta forma:

$$
\hat{\theta}_A=
\frac{1}{N}
\sum_{i=1}^{N}
\frac{f(U_i)+f(1-U_i)}{2}
$$

Como:

$$
f(x)=e^{x^2}
$$

entonces:

$$
\hat{\theta}_A=
\frac{1}{N}
\sum_{i=1}^{N}
\frac{e^{U_i^2}+e^{(1-U_i)^2}}{2}
$$

```python
def antitetica(f,N):
    s=0

    for i in range(N):
        U = r.random()
        s += (f(U) + f(1-U))/2

    return s/N
```

```python
l3 = []

for i in range(100):
    M = antitetica(f,1000)
    l3.append(M)

print(np.mean(l3))
print(np.var(l3))
```

```python
print("Monte Carlo crudo")
print("Media:", np.mean(l1))
print("Varianza:", np.var(l1))

print("Estimador con f(1-U)")
print("Media:", np.mean(l2))
print("Varianza:", np.var(l2))

print("Estimador antitético")
print("Media:", np.mean(l3))
print("Varianza:", np.var(l3))
```

```python
reduccion = (1 - np.var(l3)/np.var(l1))*100

print("Reducción porcentual de varianza:", reduccion, "%")
```

# Interpretación de resultados

Se estimó la integral

$$
\int_0^1 e^{x^2}dx
$$

utilizando tres procedimientos diferentes.

Primero se aplicó Monte Carlo crudo usando únicamente valores aleatorios de la forma

$$
f(U).
$$

Después se utilizó la variable antitética

$$
f(1-U).
$$

Finalmente se construyó el estimador antitético combinando ambos valores:

$$
\frac{f(U)+f(1-U)}{2}.
$$

Los resultados muestran que los tres estimadores producen valores muy similares para la integral.

Sin embargo, al comparar las varianzas obtenidas, se observa que el estimador antitético presenta una variabilidad mucho menor que la del método de Monte Carlo crudo.

En esta simulación se obtuvo una reducción aproximada de:

$$
88.7\%
$$

en la varianza.

Por lo tanto, el uso de variables antitéticas permite obtener una estimación más estable y precisa de la integral

$$
\int_0^1 e^{x^2}dx,
$$

manteniendo el mismo número de simulaciones.

# Problema 3. Técnicas de remuestreo: Bootstrap y Jackknife

Se aplican las técnicas de remuestreo Bootstrap y Jackknife al ejemplo de clase:

$$
\{35,42,38,40,45,37,39,41,44,36,43,40\}
$$

El objetivo es estimar la media de los datos y analizar la variabilidad del estimador.

Sea:

$$
\bar{x}=\frac{1}{n}\sum_{i=1}^{n}x_i
$$

```python
import numpy as np
import random as r
```

```python
datos = [35, 42, 38, 40, 45, 37, 39, 41, 44, 36, 43, 40]

n = len(datos)

print(datos)
print(n)
```

```python
def media(datos):
    s = 0

    for i in range(len(datos)):
        s += datos[i]

    return s/len(datos)
```

```python
media_original = media(datos)

print("Media original:", media_original)
```

## Bootstrap

Bootstrap consiste en formar nuevas muestras del mismo tamaño que la muestra original, seleccionando datos **con reemplazo**.

Para cada muestra bootstrap se calcula:

$$
\bar{x}^{*}=\frac{1}{n}\sum_{i=1}^{n}x_i^{*}
$$

```python
def bootstrap(datos):
    muestra = []

    n = len(datos)

    for i in range(n):
        k = r.randint(0,n-1)
        muestra.append(datos[k])

    return media(muestra)
```

```python
l_boot = []

for i in range(1000):
    M = bootstrap(datos)
    l_boot.append(M)

print(np.mean(l_boot))
print(np.var(l_boot))
```

```python
plt.hist(l_boot, bins=20)
plt.title("Distribucion Bootstrap")
plt.xlabel("Media")
plt.ylabel("Frecuencia")
plt.show()
```

## Jackknife

Jackknife consiste en eliminar una observación a la vez.

Como la muestra tiene \(n=12\) datos, se forman 12 muestras diferentes.

Para cada muestra se calcula:

$$
\bar{x}_{(i)}
=
\frac{1}{n-1}
\sum_{j\ne i}x_j
$$

```python
def jackknife(datos):
    l = []

    n = len(datos)

    for i in range(n):
        muestra = []

        for j in range(n):
            if j != i:
                muestra.append(datos[j])

        M = media(muestra)
        l.append(M)

    return l
```

```python
l_jack = jackknife(datos)

print(l_jack)
print(np.mean(l_jack))
print(np.var(l_jack))
```

```python
plt.hist(l_jack, bins=10)
plt.title("Distribucion Jackknife")
plt.xlabel("Media")
plt.ylabel("Frecuencia")
plt.show()
```

En Bootstrap la variabilidad es mayor porque se generan muchas muestras aleatorias con reemplazo. En Jackknife la variabilidad es menor porque solo se elimina una observación a la vez. Ambos métodos dan una media cercana a la media original.

## Comparación

```python
print("Media original:", media_original)

print("Bootstrap")
print("Media:", np.mean(l_boot))
print("Varianza:", np.var(l_boot))

print("Jackknife")
print("Media:", np.mean(l_jack))
print("Varianza:", np.var(l_jack))
```

## Interpretación de resultados

La media original se calcula directamente con los datos dados.

En Bootstrap se generan muchas muestras con reemplazo. Por eso, algunos datos pueden repetirse y otros pueden no aparecer en una muestra.

En Jackknife se elimina un dato a la vez y se vuelve a calcular la media.

Los valores obtenidos con Bootstrap y Jackknife son cercanos a la media original, pero permiten observar la variabilidad del estimador.
