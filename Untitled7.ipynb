# Distribución de Poisson

## Introducción

En esta práctica se trabaja con la distribución de Poisson, que es una distribución de probabilidad discreta. Esta distribución se usa para modelar el número de veces que ocurre un evento en un intervalo de tiempo, espacio o área, siempre que los eventos ocurran de manera independiente y con una tasa promedio constante.

Por ejemplo, se puede usar para representar:

- el número de llamadas que llegan a una oficina en una hora,
- el número de clientes que llegan a un banco,
- el número de errores en una página,
- el número de autos que pasan por una caseta,
- o el número de fallas en un sistema durante cierto tiempo.

La variable aleatoria de Poisson toma valores enteros no negativos:

$$
X = 0,1,2,3,\ldots
$$

Su función de probabilidad está dada por:

$$
P(X=i)=e^{-\lambda}\frac{\lambda^i}{i!}, \qquad i=0,1,2,\ldots
$$

donde $\lambda$ representa el promedio esperado de ocurrencias del evento.

En esta práctica se simulará una variable aleatoria con distribución de Poisson usando el método de la transformada inversa, como se vio en clase.



## Interpretación de la distribución de Poisson

La distribución de Poisson sirve para calcular la probabilidad de que un evento ocurra cierto número de veces dentro de un intervalo determinado.

Por ejemplo, si en promedio llegan 4 clientes por minuto, entonces podemos usar una variable aleatoria Poisson con:

$$
\lambda = 4
$$

Esto no significa que siempre llegan exactamente 4 clientes, sino que 4 es el promedio. A veces pueden llegar 2, otras veces 5, otras veces 0, etc.

Por eso esta distribución es útil en simulación, ya que permite generar valores aleatorios que representan situaciones reales donde ocurren eventos de forma discreta.

## Demostración de la identidad usada

En clase se usó la siguiente identidad para calcular las probabilidades de forma recursiva:

$$
P_{i+1}=\frac{\lambda}{i+1}P_i
$$

A continuación se demuestra esta igualdad.

Sabemos que para una variable aleatoria Poisson:

$$
P_i=P(X=i)=e^{-\lambda}\frac{\lambda^i}{i!}
$$

También:

$$
P_{i+1}=P(X=i+1)=e^{-\lambda}\frac{\lambda^{i+1}}{(i+1)!}
$$

Ahora se usa que:

$$
(i+1)!=(i+1)i!
$$

Entonces:

$$
P_{i+1}=e^{-\lambda}\frac{\lambda^{i+1}}{(i+1)i!}
$$

Como:

$$
\lambda^{i+1}=\lambda^i\lambda
$$

se puede escribir:

$$
P_{i+1}=e^{-\lambda}\frac{\lambda^i\lambda}{(i+1)i!}
$$

Separando términos:

$$
P_{i+1}=\frac{\lambda}{i+1}\left(e^{-\lambda}\frac{\lambda^i}{i!}\right)
$$

Pero:

$$
e^{-\lambda}\frac{\lambda^i}{i!}=P_i
$$

Por lo tanto:

$$
P_{i+1}=\frac{\lambda}{i+1}P_i
$$

Con esto queda demostrada la identidad usada en el algoritmo.

## Método de la transformada inversa para Poisson

Para generar una variable aleatoria con distribución Poisson se puede usar el método de la transformada inversa.

La idea es generar primero un número aleatorio:

$$
R \sim U(0,1)
$$

Después se van acumulando las probabilidades de la distribución de Poisson hasta que la suma acumulada sea mayor que el número aleatorio generado.

El algoritmo visto en clase es:

1. Generar un número aleatorio uniforme:

$$
R \sim U(0,1)
$$

2. Inicializar:

$$
i=0
$$

$$
p=e^{-\lambda}
$$

$$
F=p
$$

3. Si se cumple que:

$$
R < F
$$

entonces:

$$
X=i
$$

y se termina el procedimiento.

4. Si no se cumple, se calcula la siguiente probabilidad usando:

$$
p=\frac{\lambda}{i+1}p
$$

y se actualiza la probabilidad acumulada:

$$
F=F+p
$$

5. Después se aumenta el valor de \(i\):

$$
i=i+1
$$

y se repite el procedimiento hasta que el número aleatorio quede dentro de la probabilidad acumulada.

Este método funciona porque se compara el número aleatorio uniforme con la función de distribución acumulada. Así se puede saber qué valor de la variable aleatoria \(X\) corresponde al número generado.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

def poisson(lam):
    U = random.random()

    i = 0
    p = np.exp(-lam)
    F = p

    while U >= F:
        p = (lam * p) / (i + 1)
        F = F + p
        i = i + 1

    return i
```

```python
lam = 4
n = 1000

M = []

for j in range(n):
    X = poisson(lam)
    M.append(X)

M[:20]
```

## Resultados de la simulación

Se generaron 1000 valores aleatorios con distribución de Poisson usando:

$$
\lambda = 4
$$

Esto significa que el promedio esperado de ocurrencias es aproximadamente 4.




```python
plt.hist(M, bins=range(0, max(M)+2), density=True, edgecolor='black')
plt.title("Distribución de Poisson simulada")
plt.xlabel("Valores de X")
plt.ylabel("Frecuencia relativa")
plt.show()
```

## Conclusión

En esta práctica se pudo simular una variable aleatoria con distribución de Poisson usando el método de la transformada inversa. Para hacer el procedimiento se usó la fórmula recursiva:

$$
P_{i+1}=\frac{\lambda}{i+1}P_i
$$

Esta fórmula ayuda a calcular las probabilidades de una forma más sencilla, porque no es necesario calcular factoriales en cada paso.

También se observó que, al usar $\lambda=4$, la mayoría de los valores simulados quedaron cerca de 4. Esto era lo esperado, ya que en una distribución de Poisson el parámetro $\lambda$ representa el promedio de ocurrencias.

Con esto se puede ver que la distribución de Poisson es útil para simular situaciones donde se cuenta cuántas veces ocurre un evento en cierto intervalo.
